```python
import numpy as np
import math

def gera_sol(sol):
  for i in range(len(sol)):
    sol[i] = np.random.randint(0,2)
  return sol

def calcula_fo(b, p, s, n, cap, pena):
  fo = 0
  pt = 0
  for i in range(n):
    fo = fo + s[i]*b[i]
    pt = pt + s[i]*p[i]
  if pt > cap:
    #print('Solução inviável')
    fo = fo -pena*(pt - cap)
  return fo

def melhor_vizinho(s, b, p, n, pena):
  melhor = -math.inf
  for i in range(n):
    s[i] = 1 - s[i] #Troca o bit
    f = calcula_fo(b, p, s, n, cap, pena)
    s[i] = 1 - s[i] #destroca
    if f > melhor:
      melhor = f
      bi = i
  return melhor, bi

def busca_local(s, b, p, n, pena, fo):
  #fo = calcula_fo(b, p, s, n, cap, pena)
  #print('Solução inicial =', fo)
  flag = 1
  while flag:
    flag = 0
    melhor, bi = melhor_vizinho(s, b, p, n, pena)
    if melhor > fo:
      s[bi] = 1 - s[bi]
      fo = melhor
      flag = 1
      #print(s)
      #print(fo)
  return s, fo

def busca_local2(s, b, p, n, pena, fo):  
  flag = 1
  while flag == 1:
    flag = 0
    melhor = -10000
    for i in range(n):
      if s[i] == 1:
        for j in range(n):
          if s[j] == 0:
            s[i] = 0
            s[j] = 1
            f = calcula_fo(b, p, s, n, cap, pena)
            s[i] = 1
            s[j] = 0
            if f > melhor:
              melhor = f
              mi = i
              mj = j
    if melhor > fo:
      fo = melhor
      s[mi] = 0
      s[mj] = 1
      flag = 1
  return s, fo




#DADOS DO PROBLEMA-------------------#
n = 8 #Quantidade de objetos
beneficio = [4, 3, 2, 6, 2, 3, 5, 4]
peso =      [5, 4, 3, 9, 4, 2, 6, 7]
cap = 20 #Capacidade da mochila
#------------------------------------#
penalidade = 2
#------------------------------------#
sol = n*[0] #Vetor para guardar a Solução
sol_star = n*[0] #Vetor para guardar a Solução
sol = gera_sol(sol)
k = 1
fo_star = -1000000
fo = calcula_fo(beneficio, peso, sol, n, cap, penalidade)
while k <= 2:
  if k == 1:    
    sol, fo = busca_local(sol, beneficio, peso, n, penalidade, fo)
    #sol, fo = busca_local2(sol, beneficio, peso, n, penalidade, fo)
  if k == 2:
    #sol, fo = busca_local(sol, beneficio, peso, n, penalidade, fo)
    sol, fo = busca_local2(sol, beneficio, peso, n, penalidade, fo)    
  if fo > fo_star:
    print('Vizinhança =', k)
    print(fo)
    fo_star = fo
    for i in range(n):
      sol_star[i] = sol[i]
    k = 1 
  else:
    k = k+1
  

print('Fo final =', fo_star)
print(sol_star)
```
